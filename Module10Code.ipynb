To better understand the relationship between disability and depreviation we are going to model their relationship

```python
#investigate the relationship between explanatory var and target var with a scatter plot
plt.scatter(df['Health Deprivation and Disability Score'],df['Index of Multiple Deprivation (IMD) Score'])
plt.title('Health/Disability depreviation score VS index of multiple deprivation score')
plt.xlabel('Health/Disability Depreviation score')
plt.ylabel('IMD Score')
```

```python
#check explanatory var for outliers
sns.boxplot(df['Health Deprivation and Disability Score'])
```

```python
#we can clearly see values above 2 and below -2 are acting as outliers, so lets grab the list of offending indices

outliersa = df[df['Health Deprivation and Disability Score']>2].index.to_list()
outliersb = df[df['Health Deprivation and Disability Score']<-2].index.to_list()

#copy the df to be safe and drop the outliers
dfc = df.copy()
dfc.drop(outliersa, inplace=True)
dfc.drop(outliersb,inplace=True)
```

```python
#define  response var
y=df['Index of Multiple Deprivation (IMD) Score']
#define explanatory var
x = df['Health Deprivation and Disability Score']
#constant to predictor vars
x = sm.add_constant(x)
#fit the model
model = sm.OLS(y,x).fit()
#view model summary
print(model.summary())
```

From the output summary we can see that the fitted regression equation is:

Score = 21.6694 + 14.9374\*(health dep and disability score)

So each additional score point in the explanatory variable results in an increase in IMD score of roughly 14.94. The intercept score of 21.67 shows the average expected IMD score for a record with 0 health dep and disability score.

Interpreting the model summary:

 **P\>|t|: 0.000**

- Since this is lower than 0.5 we can say that there is a statistically significant association between the two variables

**R-squared: 0.701**  

- 70.1 percent of variation in IMD can be explained by Health Deprivation and Disability score.

```python
#Check the residuals to increase confidence in model

fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model,'Health Deprivation and Disability Score', fig=fig)
```

Since the residuals seem to be roughly centered around 0 (although not perfect), this is an indication that the heteroscedasticity is not a problem for the explanatory variable
